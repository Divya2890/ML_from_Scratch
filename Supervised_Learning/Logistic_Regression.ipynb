# Implementating Logistic Regression

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
from scipy.stats import norm
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
import numpy as np
class Logistic_Regression:

  def sigmoid(self,z):
    z = np.array(z, dtype=float)
    return 1/(1+ np.exp(-z))

  # defining parameters and hyper parameters

  def __init__(self, iterations=1000, learning_rate=0.01):
    self.iterations = iterations
    self.learning_rate = learning_rate
    self.weights = None
    self.bias = 0

  # find the predictions of the model

  def fit(self,x,y):
    no_of_rows, no_of_features = x.shape
    self.weights = np.zeros(no_of_features)
    self.bias =0
    #print(self.weights.shape,"weigths shape")
    #print(x.shape,"shape of x")
    #print("yshape",y.shape)

  # Now applying gradient descendent method to update the weights and bias. we obtain dw and db by applying derivative for the cost function which results in
    for _ in range(self.iterations):
      self.y_predictions = self.sigmoid(np.dot(x,self.weights)+self.bias)
      #print("predicitoen",self.y_predictions.shape)
      dw = (1/no_of_rows ) * np.dot(x.T,(self.y_predictions-y))
      db = (1/no_of_rows ) * np.sum((self.y_predictions-y))
      self.weights = self.weights - self.learning_rate * dw
      self.bias = self.bias - self.learning_rate * db
      #print("Weights",self.weights,"bias", self.bias)


  def predict(self, x):
    y_predictions = self.sigmoid(np.dot(x,self.weights)+self.bias)
    prediction_class = [0 if y <= 0.5 else 1 for y in y_predictions]
    return prediction_class

```

```python
from sklearn import datasets
data = datasets.load_breast_cancer()
x,y = data.data,data.target
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.3,random_state=43)
model = Logistic_Regression()
model.fit(x_train,y_train)
results = model.predict(x_test)
print("accuracy_Score",accuracy_score(results,y_test))
```
